### 1. Importing Dependencies

```python
import gym
from stable_baselines3 import A2C
from stable_baselines3.common.vec_env import VecFrameStack
from stable_baselines3.common.evaluation import evaluate_policy
from stable_baselines3.common.env_util import make_atari_env
import os
```

### 2. Test Environemnt

```python
environment_name = 'Breakout-v4'
env = gym.make('Breakout-v4', render_mode='human')
```

```python
env.reset()
```

```python
env.action_space
```

```python
env.observation_space # Image based model
```

```python
episodes = 5
for episode in range(1,episodes+1):
    obs = env.reset()
    done = False
    score = 0
    
    while not done:
        env.render()
        action = env.action_space.sample()
        obs,reward,done,info,_ = env.step(action)
        score+= reward
    print('Episode:{} Score:{}'.format(episode,score))
env.close()
```

### 3. Vectorize Environment and Train Model

```python
# make_atari_env is a helper from static baselines that help create wrapped Atari environments.
env = make_atari_env('Breakout-v4',n_envs=4,seed=0)
# VecFrameStack allows you to stack the environments together.
env = VecFrameStack(env,n_stack=4)
```

```python
env.reset()
```

```python
env.render(mode='human')
```

```python
env.close()
```

```python
log_path =os.path.join('Training','Logs')
# Because our observations are images we use CnnPolicy
model = A2C('CnnPolicy',env,verbose=1,tensorboard_log=log_path)
```

```python
model.learn(total_timesteps=100000)
```

### 4. Save and Reload Model

```python
a2c_path = os.path.join('Training','Saved Models','A2C_Breakout_Model')
model.save(a2c_path)
```

```python
del model
```

```python
env = make_atari_env('Breakout-v4',n_envs=1,seed=0)
env = VecFrameStack(env,n_stack=4)
```

```python
model = A2C.load(a2c_path,env)
```

### 5. Evaluate and Test

```python
# We can evaluate on only one environment
evaluate_policy(model,env,n_eval_episodes=10,render=True)
```

```python
obs = env.reset()
while True:
    action, _states = model.predict(obs)
    obs, rewards, dones, info = env.step(action)
    env.render()
```
